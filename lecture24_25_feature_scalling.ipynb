```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=200)  # mean=50, std=10
df = pd.DataFrame({"Feature": data})
```

```python
df.head()
```

**Feature Scalling (Standaradization)**:

```python
standardization=StandardScaler()
df["Feature_Standardization"] = standardization.fit_transform(df[["Feature"]])
```

```python
df.head()
```

** **Feature Scalling (Normalization using minmax scaller)**:
 - when you know min and max of numerical column then used normalization else standarization

*   List item



```python
scaler=MinMaxScaler()
df["Feature_Normalization"] = scaler.fit_transform(df[["Feature"]])
```
