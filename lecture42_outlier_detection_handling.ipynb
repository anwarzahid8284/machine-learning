**Outlier Detection**

```python
df[(df['cgpa']>8.8) | (df['cgpa']<5.11)]
```

```python
# so the value above from 8.80 and less from 5.11 is outlier so we trim it or we can use
# capping to handle the outlier
```

**Trimming outlier**

- remove outlier if the less number of outlier in column

```python
new_df=df[(df['cgpa']<8.8) & (df['cgpa']>5.11)]
```

```python
new_df
```

**Capping Outlier**
- upper outlier replace with uperlimit value
- lower outlier replace with lowerlimit value
- Only work if column is normally distributed

```python
upper_limit_value= df['cgpa'].mean()+3*df['cgpa'].std()
lower_limit_value= df['cgpa'].mean()-3*df['cgpa'].std()
```

```python
upper_limit_value
```

```python
df['cgpa']=np.where(df['cgpa']>upper_limit_value,upper_limit_value,np.where(df['cgpa']<lower_limit_value,lower_limit_value,df['cgpa']))
```

```python
df[(df['cgpa']>upper_limit_value) | (df['cgpa']<lower_limit_value)] # outlier is handle no outlier is found
```
